With the above summary, we can see that **the revenue column has 1,074 missing values.**

Before dealing with missing values, first, let's check and correct column with str or character type.

```python
# view unique values in sales_method & state
print(product_sales.sales_method.unique())
print(' ')
print(product_sales.state.unique())
print(' ')
```

There are no misspells in the states columns. 

However, **the sales_method has two misspells: em + call, email. It should be Email + Call & Email.** We will address that issue.

```python
# correct sales_method misspilled values
product_sales.loc[product_sales.sales_method=='em + call', 'sales_method'] = 'Email + Call'
product_sales.loc[product_sales.sales_method=='email', 'sales_method'] = 'Email'
```

```python
# a count of sales_method
product_sales.sales_method.value_counts()
```

```python
# the missing values of each column
product_sales.isna().sum()
```

```python
# find the null values in each sales_method
counts_method = product_sales.sales_method.value_counts(dropna=False)
na_revenue = product_sales.groupby('sales_method')['revenue'].count().reindex(counts_method.index)

na_counts = pd.concat([counts_method, na_revenue], axis=1).assign(NaNs=counts_method-na_revenue)
na_counts.assign(NaNs_pct=round(na_counts.NaNs/na_counts.sales_method, 2))
```

Since **each sales methods** has **less than 10% null values** in the revenue column, I'll simply remove those rows. After dropping the missing values from the revenue column, **there are 13,926 rows left.**

```python
# drop rows with missing values
product_sales = product_sales.dropna()
```

### 2: Analysis to find - What does the spread of the revenue look like overall? And for each method?

**Graph 6** shows the a univariate (one variable) distribution using kernel density estimation. Probability of getting a sale with smaller revenue is far greater than a sale with a larger revenue.

```python
# plot the density of revenue
ax = sns.kdeplot(data=product_sales, x='revenue', alpha=.5, linewidth=0, fill=True)
plt.title('Graph 6: The density of revenue overall')
plt.savefig('Graph 6: The density of revenue overall.png')
```

In **graph 7,** we see the revenue distribution in the form of a histogram for the overall as well as it highlights the distribution for each sale method

**Graph 8** has two plot about revenue and sales methods. Each tells a different story. For example, **the Email + Call** brings **more money on average then the other methods.** It's median was the highest. However, the Email method overall brought more revenue than the other methods since it was used the most.

```python
# plot two plots together (boxplot & count) for revenue and sales_methods
order = ['Email + Call', 'Call', 'Email']

fig, ax = plt.subplots(1, 2, figsize=(10,5))
sns.boxplot(data=product_sales, x='sales_method', y='revenue', ax=ax[0], order=order)

revenue_sum = product_sales.groupby('sales_method')['revenue'].sum().to_frame()
sns.barplot(x=revenue_sum.index, y=revenue_sum.revenue, ax=ax[1], order=order)

fig.suptitle('Graph 8: The sales methods and revenue')
ax[0].set_title('Boxplot')
ax[1].set_title('Sum of revenue')
ax[1].set_ylabel('')
plt.savefig('Graph 6: The sales methods and revenue.png');
```

```python
# the IQR of each method's revenue
from scipy.stats import iqr

call = product_sales.loc[product_sales.sales_method=='Call', 'revenue']
email = product_sales.loc[product_sales.sales_method=='Email', 'revenue']
email_call = product_sales.loc[product_sales.sales_method=='Email + Call', 'revenue']

print('Call IQR: ', iqr(call))
print('Email IQR: ', iqr(email))
print('Email + Call IQR: ', iqr(email_call))
```

## Part 3. Definition of Business Metric for the business to monitor and track the success of Different Sales Methods

#### Metric: 
Average Weekly Revenue by Sales Method

#### Definition: 
Total sum of revenue generated by each sales method divided by the weeks spent under each method to achieve the sales.

#### Business Goal:
To increase the revenue generated from sales by applying and monitoring the performance of different sales methods or approach (Email, Calls, Email+Calls)

#### Monitoring Approach: 
Track the average weekly revenue for each sales method over time to assess effectiveness.

       sales_method    revenue     week  avg_weekly_revenue

#### Initial Estimates based on current data:

           sales_method  avg_weekly_revenue
                  Call        37927.25
                 Email       112052.97
          Email + Call        68042.78

```python
metric_df= product_sales.groupby('sales_method').agg({'revenue':'sum','week':'max'}).reset_index()
metric_df['avg_weekly_revenue']=metric_df['revenue']/metric_df['week']
print(metric_df)
```
